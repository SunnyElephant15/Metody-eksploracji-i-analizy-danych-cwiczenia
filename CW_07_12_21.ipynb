```python
import numpy as np
import numpy.random as npr
```

```python
# N - liczba bandytów
N = 10
```

```python
# Prawdziwe wartości oczekiwane rozładów prawdopodobieństwa
# wygranej dla N maszyn:
prawdziwe_średnie = npr.normal(0, 1, N)
```

```python
prawdziwe_średnie
```

```python
# Prawdziwe odchylenia standardowe rozładów prawdopodobieństwa
# wygranej dla N maszyn:
prawdziwe_odst = np.ones(N)
prawdziwe_odst
```

```python
początkowe_zakładane_średnie = npr.normal(0, 1, N)
początkowe_zakładane_średnie
```

```python
empiryczne_średnie = początkowe_zakładane_średnie
```

```python
# Nrozgr - liczba rozgrywek w kasynie = liczba "pociagnięć wajchy"
# jednego z bandytów
Nrozgr = 10
# eps - próg przewagi eksploatacji nad eksploracja
eps = 0.1
# Liczba rozgrywek z każdą z maszyn
lrozgr = np.zeros(N)
```

```python
# (x[0] + x[1] + ... + x_[k-1]) / k = S[k]
# x[k]
# S[k+1] = (x[0] + x[1] + ... + x_[k-1] + x[k]) / (k + 1) =
# (x[0] + x[1] + ... + x_[k-1]) / (k + 1) + x_[k] / (k + 1) =
# (x[0] + x[1] + ... + x_{k-1}) / k * (k / (k+1)) + x_[k] / (k + 1) =
# S
```

```python
całkowita_wygrana = 0
k = 0
llos_tab = npr.uniform(size = Nrozgr)
while k < Nrozgr:
  llos = llos_tab[k]
  if llos > eps:
    index = np.argmax(empiryczne_średnie)
  else:
    index = npr.random_integers(0, N-1)
  print(index)
  lrozgr[index] +=1
  wygrana = npr.normal(prawdziwe_średnie[index], prawdziwe_odst[index])
  całkowita_wygrana + wygrana
# Trzeba zaktualizowac tablice 'empityczne_średnie'
  k +=1
# Narysowac wykres calkowitej wygranej po kazdym kroku
```
